```python
import os
import sys
root_folder = os.path.dirname(os.getcwd())
sys.path.append(root_folder)
from sectorizedradialprofile.calculate_radial_profile import CalculateRadialProfile
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

Load the data

```python
# data_file = 'data_2_circles.tif'
data_file = 'BC_bottom_sm_crop_0050.tif'
data = np.array(Image.open(data_file))
```

Visualize data

```python
data
```

```python
plt.figure(0)
plt.imshow(data)
```

```python
center = {'x0': 79, 'y0': 93}
angle_range = {'from': 0, 'to': 360}
radius = 40
o_profile = CalculateRadialProfile(
    data=data, center=center, radius=radius, angle_range=angle_range)
o_profile.calculate()
plt.figure(1)
plt.imshow(o_profile.working_data)
plt.show()
```

```python
profile = o_profile.radial_profile
plt.figure(2)
plt.errorbar(x=profile.index, y=profile['mean'], yerr=profile['std'], fmt='-.')
plt.show()
```
